```
!pip install Keras==2.2.5
!pip install keras_bert==0.81.1
!pip install git+https://github.com/keras-team/keras-contrib.git/@master#egg=keras-contrib
import numpy as np
import os
from keras_bert import extract_embeddings
from keras_bert import load_trained_model_from_checkpoint
import codecs
from keras_bert import Tokenizer
from keras.models import Sequential
from keras.layers import Embedding, Bidirectional, LSTM, TimeDistributed,Dense
from keras_contrib.layers import CRF
from keras.callbacks import Callback
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
from keras.models import load_model
from keras_contrib.layers.crf import CRF, crf_loss, crf_viterbi_accuracy
```

```
pretrained_path = './chinese_L-12_H-768_A-12'
config_path = os.path.join(pretrained_path, 'bert_config.json')
checkpoint_path = os.path.join(pretrained_path, 'bert_model.ckpt')
vocab_path = os.path.join(pretrained_path, 'vocab.txt')
model_path=r'./bert_ner.h5'
train_path=r'./data/train.txt'
train_less_path=r'./data/train_less.txt'
test_less_path=r'./data/test_less.txt'
test_path=r'./data/test.txt'
write_path=r'data/test_label'
relation_dic_path = r'./data/relation.txt'
EMBED_DIM = 200
BiRNN_UNITS = 200
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
```

```
#模型字典
token_dict = {}
with codecs.open(vocab_path, 'r', 'utf8') as reader:
    for line in reader:
        token = line.strip()
        token_dict[token] = len(token_dict)
tokenizer = Tokenizer(token_dict)

with open(relation_dic_path, encoding="utf-8", errors='ignore') as dic_path_file_object:
    dic_len = dic_path_file_object.read()
dic_arr = dic_len.split('\n')
dic = {}
for line in dic_arr:
    temp = line.split(' ')
    if len(temp) == 2:
        dic[temp[0]] = temp[1]

chunk_tags=len(dic_arr)*2+1

def get_key (dict, value):
    return [k for k, v in dict.items() if v == value]
```

```
#语料样本向量化
#pad_num：自定义语料squence长度
def pre_y(sentence, label, relation, pad_num):
    y=np.zeros((len(sentence),pad_num))
    for i in range(len(sentence)):
        sen = sentence[i]
        tokens = tokenizer.tokenize(sen)
        lab = label[i]
        relat = relation[i]
        lab_y = np.zeros(len(tokens))
        if lab!='':
            top = None
            #label有多项
            t=[]
            n=[]
            for u in lab.split(','):
                tokens_lab = tokenizer.tokenize(u)
                num_lab = len(tokens_lab) - 2
                for j in range(len(tokens)):
                    if j + num_lab - 1 < len(tokens):
                        if (tokens[j] == tokens_lab[1]) & (tokens[j + num_lab - 1] == tokens_lab[1 + num_lab - 1]):
                            t.append(j)
                            n.append(num_lab)
            if len(t)==0:
                print('error:' + sen)
                break
            for t_num in range(len(t)):
                re_tag = dic[relat]
                lab_y[t[t_num]] = int(re_tag)*2+1  # B-COM-relation
                for g in range(t[t_num]+1,t[t_num]+n[t_num]):
                    lab_y[g] = int(re_tag)*2+2  # I-COM-relation
        y[i]=np.lib.pad(lab_y, (0,pad_num-len(tokens)), 'constant', constant_values=(0,0))
    return y.reshape((len(sentence),pad_num,1))
```

```
#语料标签向量化
#pad_num：自定义语料squence长度
def pre_x(sentence,pad_num):
    x = extract_embeddings(pretrained_path, sentence)
    #padding
    x_train = np.zeros((len(sentence), pad_num, x[0].shape[1]))
    for i in range(len(sentence)):
        for j in range(len(x[i])):
            if len(x[i]) > pad_num:
                print('error:超出范围！'+str(len(x[i]))+str(sentence[i]))
                break
            x_train[i][j] = x[i][j]
    return x_train
```

```

#训练模型
with open(train_less_path, encoding="utf-8", errors='ignore') as sentence_file_object:
  sentence = sentence_file_object.read()
  s = sentence.split('\n')
  label = []
  relation = []
  sen = []
  for item in s:
      arr = item.split('\t')
      if len(arr) == 4:
          label.append(arr[0]+','+arr[1])
          relation.append(arr[2])
          sen.append(arr[3])

  y = pre_y(sen, label, relation, 128)
  x = pre_x(sen, 128)
```

```
from keras_contrib.metrics.crf_accuracies import _get_accuracy
def create_joined_crf_accuracy(crf):
    def accuracy(y_true, y_pred):        
        X = crf.input
        mask = crf.input_mask
        y_pred = crf.viterbi_decoding(X, mask)
        return _get_accuracy(y_true, y_pred, mask, crf.sparse_target)

    return accuracy
def create_joined_crf_loss(crf):
    def loss(y_true, y_pred):
        offset = 0
        
        X = crf.input
        mask = crf.input_mask
        nloglik = crf.get_negative_log_likelihood(y_true, X, mask)
        return nloglik

    return loss
```

```
from keras.layers import Input, Lambda
from keras_contrib import losses,metrics
model = Sequential()

# lstm = LSTM(BiRNN_UNITS // 2,return_sequences=True)
# lstm =Lambda(LSTM)(BiRNN_UNITS // 2)
# lstm =Lambda(lambda x:x(BiRNN_UNITS//2,return_sequences=True))(LSTM)
model.add(Bidirectional(LSTM(BiRNN_UNITS // 2,return_sequences=True)))
# model.add(TimeDistributed(Dense(len(chunk_tags))))
# crf = CRF(chunk_tags, sparse_target=True)
# crf = Lambda(CRF)(chunk_tags)
# crf = Lambda(lambda x:x(chunk_tags, sparse_target=True))(CRF)
model.add(CRF(chunk_tags, sparse_target=True))
model.compile('adam', loss=crf_loss, metrics=[crf_viterbi_accuracy])
# model.compile(loss=create_joined_crf_loss(crf), optimizer='adam', metrics=[crf_viterbi_accuracy])
```

```
def predict(test_path,model_path,write_path):
    with open(test_path, encoding="utf-8", errors='ignore') as test_path_file_object:
        test = test_path_file_object.read()
    test_s = test.split('\n')
    test_x = pre_x(test_s, 128)
    model=load_model(model_path, custom_objects={"CRF": CRF, 'crf_loss': crf_loss,'crf_viterbi_accuracy': crf_viterbi_accuracy})
    pre = model.predict(test_x)
    result = ''
    for i in range(len(pre)):
        flag = tokenizer.tokenize(test_s[i])
        flag_pre = [np.argmax(j) for j in pre[i]][:len(flag)]
        com = ''
        rel = 0
        for s, t in zip(flag, flag_pre):
            if t > 0:
                if t%2 == 1:
                    rel = t//2
                    com+=','+s
                else:
                    com +=s
        if com != '':
            if com[0] == ',':
                com = com[1:]
        com = com.replace('#', '')
        if i == len(pre) - 1:
            result += com
        else:
            relname = get_key(dic, str(rel))
            result += com +','+ relname[0] + '\n'
    with open(write_path, 'a', encoding='utf-8') as file_handle:  # .txt可以不自己新建,代码会自动新建
        file_handle.write(result)
predict(test_less_path, model_path, write_path)
```
